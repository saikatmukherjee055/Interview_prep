```python
# Q: Create a DataFrame with columns A, B, and C, and fill it with random values.

import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(5, 3), columns=['A', 'B', 'C'])
df
```

```python
#Q: Extract the digits from a string column.

import pandas as pd

# Assuming df has 5 rows
df = pd.DataFrame({
    'X': [1, 2, 3, 4, 5],
    'Y': [10, 20, 30, 40, 50],
    'Z': [100, 200, 300, 400, 500]
})

# Adding string column with same length as df
df['string_col'] = ['abc123', 'def456', 'ghi789', 'jkl012', 'mno345']

print(df)

# Extracting digits
df['digits'] = df['string_col'].str.extract('(\d+)', expand=False)
df

```

```python
#Q: Extract all digits from a string column where the string can contain both single digits and groups of digits.

import pandas as pd

# Assuming df has 5 rows
df = pd.DataFrame({
    'X': [1, 2, 3, 4, 5],
    'Y': [10, 20, 30, 40, 50],
    'Z': [100, 200, 300, 400, 500]
})

# Adding string column with same length as df
df['string_col'] = ['a2b123', 'c45d67', 'e89f0', 'g123h456', 'i789j']

# Extracting all digits and joining them into a single string
df['all_digits'] = df['string_col'].str.findall('(\d+)').apply(lambda x: ''.join(x))
df
```

```python
#Q: Convert a timezone-aware datetime column to another timezone.

import pandas as pd

# Create a DataFrame with a datetime column
df = pd.DataFrame({
    'date': pd.date_range('2023-01-01', periods=5, freq='D')
})

# Localize the datetime column to UTC
df['date'] = df['date'].dt.tz_localize('UTC')

# Convert the datetime column to another timezone, e.g., US/Eastern
df['date_eastern'] = df['date'].dt.tz_convert('US/Eastern')

df

```
